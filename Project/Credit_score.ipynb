3.   个贷产品中的风险   
  * 风险类型
      * 利率风险：利率水平的变动会影响金融产品的定价。因利率波动对信贷产品定价产生的损失机利率风险
      * 流动性风险：信贷机构在某个时刻没有足够的资金量来满足客户体现需求的风险。如果缺乏流动性管理，一旦资金断裂会引发支付危机
      * 信用风险：借款人在合约到期时没有完全旅行其义务的风险。就是时间到了但是还没有换钱，或者是没还够
          * 有两大因素构成 (信用资质=还款意愿×还款能力)
              * 借贷人的还款意愿
              * 借贷人的还款能力
          * 信用风险通常贯穿在信贷产品的整个生命周期
              * 贷前审核：过滤掉低资质客户，将中、高资质客户进行放款
              * 贷中监控：放款后，持续监控信贷客群
              * 逾期催收：发生逾期后，催促还款
          * 信用风控的两大手段
              * 策略规则：根据信贷客群的申请资料或者行为表现，在一定规则下实施风险管理
              * 量化模型：建立统计学或者其他算法模型，根据模型结果实施风险管理。

      * 信贷有三个环节，我们对这三个环节都要进行风险预测
          * 贷前申请：申请评分，预测放款后是否会有违约
          * 贷中监控：行为评分，预测还款结束前未来一段时间内是否会有违约
          * 逾期催收：催收评分，预测是否会成为不良资产

      * 三大评分模型特点
          * 申请评分：考虑个人信息(比如说性别，年龄，教育经历啊等)，征信信息(在过去的借款行为啊，还款行为啊之类的)，收入信息(流水啊，房产啊，交税啊之类的)。评分结果用于制定准入策略，可分为拒绝、准入(这就是二分类)。或者是拒绝、再审核、准入等状态(再加一个分类)
          * 行为评分：考虑账龄，还款行为，消费行为，抵押品估值(抵押的是什么呢?)，逾期行为等。行为评分并不适合所有的信贷，它是适合还款周期长的产品(如房贷、车贷等)或者是循环授信类产品(如信用卡), 结果可用于提前预警，风险计量等。
          * 催收评分：考虑个人信息，账龄，还款行为，欠款余额等。评分结果可用于制定催收策略。评分结果较好的逾期人群可进行轻催收，不好的就要进行重催收





## Unit 2 评分卡模型
在构建评分卡模型的工作中，数据预处理工作和特征构建工作是至关重要的一步。数据的预处理工作可以有效处理缺失值与异常值，从而增强模型的稳健性。而特征构建工作则可以将信息从字段中加以提炼，形成有业务含义的优异特征。



1.   评分卡模型简介

  * 风控场景中的评分卡
      * 以分数的形式来衡量风险几率的一种手段
      * 是对未来一段时间内违约、逾期、失联概率的预测
      * 有一个明确的(正)区间
      * 通常分数越高越安全
      * 数据驱动
      * 风控模型呢，有如下几个评分卡类型：
          * 反欺诈评分卡
          * 申请评分卡(Application scorecard)：就是所谓的A卡。用在贷前审核环节，评估放贷后是否会违约的模型。常用特征：个人信息(年龄，性别，工作等)，央行征信信息，申请行为信息，其他辅助信息(运营商信息啊，电商信息啊之类的)
          * 行为评分卡(Behavioral scorecard)：就是所谓的B卡。用在贷后监控环节，做早期预警的工作(包括巴塞尔2.5(The Basel)及之前的AIRB的要求). 常用特征：贷后的还款行为，消费行为。通常适用于还款周期长的产品或者循环授信类产品
          * 催收评分卡(Collection scorecard)：就是所谓的C卡。用在发生逾期后的管理环节，为催收工作提供指导。催收评分卡又可细分为预测失联的失联评分卡、预测逾期加重的滚动率评分卡和预测催收后的还款率的还款评分卡。常用特征：个人信息、贷后的还款行为、消费行为、联系人信息等
  * 评分卡模型开发步骤

3.   特征构建的方法
    在评分卡模型的开发中，特征构造是及其关键的步骤。作用是将分散在不同字段中的信息加以组合，从而提炼出有价值的信息、有用的信息。再把这些作为新的特征来进行模型开发。这里举个例子来说，假设比平均速度，A跑200米，用了26秒，B跑100米，用了12秒。做一个模型时，我们就需要考虑我们需要什么特征，假如需要平均速度作为特征，那么我们就需要将"米/s"从而得到新的特征。

    所以，我们得到数据后就要去看它有什么features，每个features有什么作用，需要怎么构造新的特征。比如说，看到时间或日期，可以考虑时间的跨度啊之类的，哪段时间人最多啊之类的，或者是在这段时间内什么操作的频率啊之类的
    ![WeChat Screenshot_20210912211548](https://user-images.githubusercontent.com/68700549/133010411-041b7774-36a8-46dc-aeca-44dde111436f.png)

    部分常用的特征构造方法有
    * 求和：例如过去一段时间内的每月网购金额总和
    * 比例：例如申请贷款的月还款与月收入的占比
    * 频率：例如过去一段时间内的境外消费次数
    * 平均：例如过去一段时间内平均每次信用卡取现额度
    好的特征需要具备以下优势
      * 稳定性高
      * 区分度高
      * 差异性大
      * 符合业务逻辑

4.    数据的质量检验与处理

    *  数据集中度

        在信用风控模型的开发中，数据集中度是常见的问题。即在变量中，某单一数值的占比就占了全部样本值的绝大多数。例如，在一批训练样本中，学历为本科的样本就占了全部样本的90%。具有极高的极高的集中度的字段或者变量，需要按照风险程度进行区分。(同理，在其他数据集中也会出现这种情况，我们根据target来进行区分)
        * “多数值”与“少数值”对应的坏样本率没有显著的差别，由于包含的信息较少，对模型的开发没有太大的价值，而且往往“少数值”的产生是由于误差或者噪声，我们可以直接将字段删除
        * “多数值”与“少数值”有显著差别，且“少数值”的坏样本率低于“多数值”。此时，尽管二者的差别很显著，但是由于我们更加关注风险度高的一组，所以“少数值”的存在并不会带来额外的意义，此时也是可以直接将字段删除的
        * “多数值”与“少数值”有显著差别，且“少数值”的坏样本率高于“多数值”。此时，“少数值”的存在表明该值对应的风险很高，字段需要保留。
    
        可是，我们应该以一个怎样的界限来说明是高还是低呢?

        首先我们要计算坏样本率，将二者的比值的对数作为衡量。比如说，当A为B的10倍时，取对数后为3.3219.当A为B的十分之一是，取对数后为-3.3219. 这也就是我们为什么要取对数的原因，因为对数转换可以将互为倒数的两组数变为互为相反数。此时，我们就想，那我们是用哪个feature好呢？我们就以某个feature(它的多数值占比超过90%的那个feature为基准)，算出一个对数比例之后，然后对其他字段也检查这个比例，如果都在刚刚算出的比例之内，那么，说明这个“少数值”的存在并不会对风险产生太大的影响，可以删除。
    * 数据缺失(Missing data)

        数据缺失是非常常见的现象，我们要从两个维度去检查数据的缺失度
        * 字段维度：所谓字段，就是feature啦。就是某个feature上的缺失数据占全部样本的占比是多少
        * 样本维度：就是一个样本，假设有100个features，缺失了20个数据，我们就要算这条样本的缺失值占比

        一般来说，字段维度的缺失程度会大于样本维度的缺失程度

        缺失值处理

        https://zhuanlan.zhihu.com/p/137175585

          * 舍弃该字段或该条记录，缺失占比太高
          * 补缺：缺失占比不高，可用均值法，众数法，回归法等。补缺的方法是看变量类型。如果是数值型，可以用均值法(完全随机缺失)，回归法(针对随机缺失)，抽样法(完全随机缺失)来补缺。如果是categorical data，那么就用抽样法，众数法来补缺
          * 作为特殊值，将缺失看成一种特殊值
        在信贷评分模型中，数据的缺失可能包含着多重意义。很多时候是完全非随机缺失(missing completely at random)，其缺失状态有着业务含义
    * Outliers

        https://www.zhihu.com/question/38066650

        与缺失值类似，outliers会对模型产生一定的干扰。outliers的判断通常有聚类法，分位点法等，处理方法有删除法，替换法。我们要去学会怎么detect outliers。
      
    * 数据含义一致性

          在数据录入时，可能会把属于同一含义的记录值录成不同的表达方式，比如说"QQ"和“qQ”应该要一样，"Idnumber"和“idNumerber”要一样。还有电话号码"+8613xxx"和“13xxx”应该要一样。我们要进行统一大小写啊，国家电话号码区号之类的

For Login_Info dataset, there are two dates, one is Login date and another one is listing date. So we can calculate the difference between these two dates. Before we do that, we need to make this two columns become datetime objects. Then we can use it to calculate the difference. 

```python
## if we add column wrongly, then we can drop it
## df.drop(column=[''])

## add one column, this data column is same as LoginInfo3, but the new column is datetime object
df_PPD_LogInfo['LogInfo'] = df_PPD_LogInfo['LogInfo3'].map(lambda x : datetime.strptime(x, "%Y-%m-%d"))
## add one column, this data column is same as Listinginfo1, but the new column is datetime object
df_PPD_LogInfo['Listinginfo'] = df_PPD_LogInfo['Listinginfo1'].map(lambda x : datetime.strptime(x, "%Y-%m-%d"))

## After we convert them to datetime object, we can calculate the date difference
## Here, we can add a new column to record the date_gap
df_PPD_LogInfo['date_gap'] = df_PPD_LogInfo[['LogInfo', 'Listinginfo']].apply(lambda x : (x[1]-x[0]).days, axis=1)
```

```python
plt.hist(df_PPD_LogInfo['date_gap'], bins=200)
plt.title("the gap between login date and listing date")
```

We can also visualize the frequence too. So, for example, how many gaps is less than 30 days? how many gaps is less than 60 days? So, we can visualize it too. Here, we can set to [30, 360]. We can write a funciton to achieve it

```python
## This funciton is used to calculate the days frequency
def calculateDateGapFrequence(data, column_name, range):
  ## the format can be for example, {30: 100, 60: 200} 
  fre = {}
  for gap in range:
    freq = sum(data[column_name].apply(lambda x : x < gap))
    fre[gap] = freq
  return fre
```

Let's visualize the df_PPD_LogInfo date gap. We can use bar chart to visualize it. From this bar chart, we can know that most of them are less than 180 days

```python
date_gap_fre_of_df_PPD_LogInfo = calculateDateGapFrequence(df_PPD_LogInfo, 'date_gap', np.arange(30,361,30))
date_gap = list(date_gap_fre_of_df_PPD_LogInfo.keys())
gap = list(date_gap_fre_of_df_PPD_LogInfo.values())
plt.bar(date_gap, gap, color ='blue', width = 4)
plt.xlabel("date_gap")
plt.ylabel("gap")
plt.title("the frequency of difference between login date and listing date in PPD_LogInfo")
```
