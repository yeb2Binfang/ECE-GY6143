信用评分卡模型是最常见的金融风控手段之一，它是指根据客户的各种属性和行为数据，利用一定的信用评分模型，对客户进行信用评分，据此决定是否给予授信以及授信的额度和利率，从而识别和减少在金融交易中存在的交易风险。 评分卡模型在不同的业务阶段体现的方式和功能也不一样。

3.   个贷产品中的风险   
  * 风险类型
      * 利率风险：利率水平的变动会影响金融产品的定价。因利率波动对信贷产品定价产生的损失机利率风险
      * 流动性风险：信贷机构在某个时刻没有足够的资金量来满足客户体现需求的风险。如果缺乏流动性管理，一旦资金断裂会引发支付危机
      * 信用风险：借款人在合约到期时没有完全旅行其义务的风险。就是时间到了但是还没有换钱，或者是没还够
          * 有两大因素构成 (信用资质=还款意愿×还款能力)
              * 借贷人的还款意愿
              * 借贷人的还款能力
          * 信用风险通常贯穿在信贷产品的整个生命周期
              * 贷前审核：过滤掉低资质客户，将中、高资质客户进行放款
              * 贷中监控：放款后，持续监控信贷客群
              * 逾期催收：发生逾期后，催促还款
          * 信用风控的两大手段
              * 策略规则：根据信贷客群的申请资料或者行为表现，在一定规则下实施风险管理
              * 量化模型：建立统计学或者其他算法模型，根据模型结果实施风险管理。
          * 信贷风控流程(以贷前审核为例),如下图所示
          <img src="https://user-images.githubusercontent.com/68700549/133000149-f1a51cea-b585-4632-885a-b531dfbabe21.png" alt="WeChat Screenshot_20210912152358" style="zoom:67%;" />

      * 信息安全风险：借贷系统及其赖以运行的基础网络、处理的数据和信息由于存在的硬软件缺陷、系统集成缺陷等，以及信息安全管理中的薄弱环节而导致的安全风险。最典型的就是黑客
      * 操作风险：人为失误啊，不完备的程序控制啊，欺诈啊，犯罪活动啊。由于技术缺陷和系统崩溃引起的。
      * 欺诈风险：借款人故意隐瞒信息，虚构事实骗取贷款，并且因“主观上不想”和“客观上不能”偿还，可能造成出借人风险损失。
          * 主要有三个特征
              * 骗取贷款为最终目的
              * 使用隐瞒，虚构等欺诈性手段
              * 借款时就没打算还，包括有能力没意愿的，也包括有意愿没能力还的

          * 欺诈风险主要发生环节
              * 申请端欺诈(在申请的时候就进行欺诈)
              * 交易环节欺诈(盗卡啊，盗刷啊 之类的)
          * 申请反欺诈的主要步骤
              * 传统步骤
                  * 查黑名单(法院失信，被执行等，金融机构共享黑名单)
                  * 查多头借贷(人行征信，第三方大数据公司名单)
                  * 勾稽规则(申请资料的对比与核实)
                  * 查社交关系(通话记录啊，查是否跟黑产人群或者催收公司有联系)
                  * 验证逻辑(违反常识的信息啊)
                  * 综合考量(总体风险把控)
              * 新方法

              在互金领域，我们现在可以用大数据和生物方法来作为传统方法的补充和增强
                  * 生物方法
                      * 人脸识别
                      * 虹膜识别
                      * 基于人脸识别的活体检测
                  * 大数据方法
                      * 对用户在客户端的行为轨迹分析和对业务API的访问，区分正常用户和恶意用户
                      * 通过对用户访问数据的统计分析，发现异常访问，提供各个层面的风险预警
                      * 通过对用户访问内容的分析，基于设备指纹或用户ID关联，赋予用户标签，形成用户画像
                  * 第三方服务
                      * 整体的风控解决方案提供商、外包商：这类供应商提供风控基础架构，并根据客户需求做一定的定制化开发。一些供应商还担任集成商的角色，负责集成各种服务
                      * 风控建模外包：有一些专业团队提供专门领域的风控建模服务，帮助数据分析能力不足的互金企业解决紧急的风控建模问题
                      * 数据提供商：第三方数据提供商一直是风控数据的重要来源，比如说，电信运营商，电商平台等
                      * 各领域解决方案提供商：这类供应商常常以SaaS(Software as a service)服务(SaaS模式就是软件公司实现把所有的软件相关的工作都归类好了，用户过来直接挑选自己需要的服务，SaaS是云端部署，按需消费)的方式来提供个领域解决方案，如短信验证，人脸识别，生物探测，设备指纹等

  * 信用评分模型
      * 信贷有三个环节，我们对这三个环节都要进行风险预测
          * 贷前申请：申请评分，预测放款后是否会有违约
          * 贷中监控：行为评分，预测还款结束前未来一段时间内是否会有违约
          * 逾期催收：催收评分，预测是否会成为不良资产

      * 三大评分模型特点
          * 申请评分：考虑个人信息(比如说性别，年龄，教育经历啊等)，征信信息(在过去的借款行为啊，还款行为啊之类的)，收入信息(流水啊，房产啊，交税啊之类的)。评分结果用于制定准入策略，可分为拒绝、准入(这就是二分类)。或者是拒绝、再审核、准入等状态(再加一个分类)
          * 行为评分：考虑账龄，还款行为，消费行为，抵押品估值(抵押的是什么呢?)，逾期行为等。行为评分并不适合所有的信贷，它是适合还款周期长的产品(如房贷、车贷等)或者是循环授信类产品(如信用卡), 结果可用于提前预警，风险计量等。
          * 催收评分：考虑个人信息，账龄，还款行为，欠款余额等。评分结果可用于制定催收策略。评分结果较好的逾期人群可进行轻催收，不好的就要进行重催收
